# Python libraries Used

# Read data from the Excel file

```python
import pandas as pd
```

```python
excel_file = 'movies.xls'
movies = pd.read_excel(excel_file)
```

```python
movies.head()
```

# Read Data Sheet wise

```python
movies_sheet1 = pd.read_excel(excel_file, sheet_name=0, index_col=0)
movies_sheet1.head()
```

```python
movies_sheet2 = pd.read_excel(excel_file, sheet_name=1, index_col=0)
movies_sheet2.head()
```

```python
movies_sheet3 = pd.read_excel(excel_file, sheet_name=2, index_col=0)
movies_sheet3.head()
```

# Create single DataFrame from all the three DataFrames using concat

```python
movies = pd.concat([movies_sheet1, movies_sheet2, movies_sheet3])
```

```python
movies.shape
```

# Using the ExcelFile class to read multiple sheets

```python
xlsx = pd.ExcelFile('movies.xls')
movies_sheets = []
for sheet in xlsx.sheet_names:
    movies_sheets.append(xlsx.parse(sheet))
movies = pd.concat(movies_sheets)
```

```python
movies.shape
```

# Applying formulas on the columns

```python
movies["Net Earnings"] = movies["Gross Earnings"] - movies["Budget"]
```

# use the sot_values method to sort the data by the new column we created and visualize the top 10 movies by Net Earnings.

```python
sorted_movies = movies[['Net Earnings']].sort_values(['Net Earnings'], ascending=[False])
sorted_movies.head(10)['Net Earnings'].plot.barh()
plt.show()
```

# Pivot Table in pandas

choosing Year as the index column and Gross Earnings as the summarization column and creating a separate DataFrame from this data.

```python
movies_subset = movies[['Year', 'Gross Earnings']]
movies_subset.head()
```

call pivot_table on this subset of data. The method pivot_table takes a parameter index. As mentioned, we want to use Year as the index.

```python
earnings_by_year = movies_subset.pivot_table(index=['Year'])
earnings_by_year.head()
```

use pivot table to create some data visualizations. We can call the plot method on the DataFrame to create a line plot and call the show method to display the plot in the notebook.

```python
earnings_by_year.plot()
plt.show()
```

create another DataFrame subset but this time we will choose the columns, Country, Language and Gross Earnings.

```python
movies_subset = movies[['Country', 'Language', 'Gross Earnings']]
movies_subset.head()
```

```python
earnings_by_co_lang = movies_subset.pivot_table(index=['Country', 'Language'])
earnings_by_co_lang.head()
```

# visualize this pivot table with a bar plot

```python
earnings_by_co_lang.head(20).plot(kind='bar', figsize=(20,8))
plt.show()
```
